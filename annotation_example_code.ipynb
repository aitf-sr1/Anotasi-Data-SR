# Script Perhitungan Reliabilitas Annotator (Emosi Siswa)

Cell di bawah ini digunakan untuk menghitung seberapa sepakat dua orang annotator dalam menilai emosi siswa. 

Karena label emosi yang kita gunakan adalah **data bertingkat** (level 0 sampai 3), kita menggunakan metrik **Quadratic Weighted Kappa (QWK)**. Metrik ini lebih cerdas dari akurasi biasa karena bisa membedakan mana salah tebak yang "wajar" (cuma meleset 1 level) dan mana salah tebak yang "fatal" (meleset sampai 3 level). Kesalahan fatal akan membuat skor QWK turun drastis.

```python
import numpy as np
from sklearn.metrics import cohen_kappa_score

def hitung_kesepakatan_emosi(data_annotator_1, data_annotator_2):
    """
    Menghitung skor kesepakatan (QWK) untuk masing-masing emosi.

    Format data tiap video: [Confusion, Engagement, Frustration, Boredom]
    Level emosi (bertingkat): 0 (Sangat Rendah), 1 (Rendah), 2 (Tinggi), 3 (Sangat Tinggi)
    """
    daftar_emosi = ["Confusion", "Engagement", "Frustration", "Boredom"]
    hasil_skor = {}

    for i, emosi in enumerate(daftar_emosi):
        # Mengambil nilai khusus untuk satu emosi dari semua video
        skor_1 = [video[i] for video in data_annotator_1]
        skor_2 = [video[i] for video in data_annotator_2]

        # Menghitung jarak kesepakatannya (Weighted Kappa)
        qwk = cohen_kappa_score(skor_1, skor_2, weights='quadratic')
        hasil_skor[emosi] = qwk

    return hasil_skor

# --- CONTOH PENGGUNAAN ---
# Misal ini adalah data hasil nonton 5 video yang sama
# Urutan isi list: [Confusion, Engagement, Frustration, Boredom]

annotator_A = [
    [0, 2, 0, 1], # Video 1
    [1, 1, 0, 2], # Video 2
    [0, 3, 1, 0], # Video 3
    [2, 2, 2, 1], # Video 4
    [0, 0, 0, 3]  # Video 5
]

annotator_B = [
    [0, 3, 0, 1], # Video 1
    [1, 1, 1, 2], # Video 2
    [0, 3, 0, 0], # Video 3
    [1, 2, 2, 2], # Video 4
    [0, 1, 0, 3]  # Video 5
]

print("Skor Kesepakatan Antar-Annotator (Makin dekat ke 1 makin bagus):\n")
hasil = hitung_kesepakatan_emosi(annotator_A, annotator_B)

for emosi, skor in hasil.items():
    print(f"{emosi.ljust(15)}: {skor:.4f}")
```
